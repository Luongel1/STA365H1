## Q3
Slice sampling is a Monte Carlo Markov Chain (MCMC) method that generates samples from a given probability distribution by iteratively defining a "slice" beneath its probability density function. Instead of directly drawing samples, the method introduces an auxiliary variable $y$, which follows a uniform distribution between 0 and the function’s value at $x$, denoted as:

$$y \sim U(0, f(x))$$

This creates a horizontal slice where $f(x) > y$, restricting the valid sampling region.

The function `slice_f_at_y` identifies an interval $[x_l, x_r]$ in which $f(x)$ exceeds $y$, ensuring that:

$f(x_l) < y, \quad f(x_r) < y$

A new sample $x$ is then drawn uniformly from this interval. If this $x$ satisfies the slice condition $( f(x) > y )$, it is accepted; otherwise, the interval is iteratively narrowed until an acceptable sample is obtained.

By repeating this process, the method constructs a Markov chain whose distribution converges to the desired target distribution.


In a Metropolis within Gibbs sampling framework, Gibbs sampling is used to sequentially update parameters based on their respective full conditional distributions. However, when the full conditional is known only up to a normalizing constant, direct sampling becomes more challenging.

In such cases, instead of using Metropolis-Hastings to propose and accept/reject new values, slice sampling serves as an alternative approach. This method dynamically defines a region for sampling, avoiding the need for a proposal distribution. Instead of computing an acceptance probability, slice sampling introduces an auxiliary variable $y$ as described earlier. The algorithm then iterates until a valid sample is drawn.

In this approach, the underlying curve represents the \textbf{un-normalized full conditional probability distribution}, which typically corresponds to a posterior distribution known only up to a normalizing constant. The initial slice selection is arbitrary and determined by the user at the start of the algorithm.

```python
import numpy as np
from scipy import stats

def gibbs_with_slice_sampling(full_conditional, x_init, num_iters,
                              x_range = (0, 1), num_grid = 100):
    samples = np.zeros(num_iters)
    x = x_init

    x_grid = np.linspace(x_range[0], x_range[1], num_grid)

    for i in range(num_iters):
        y = np.random.uniform(0, full_conditional(x))

        valid_x = x_grid[full_conditional(x_grid) > y]
        if len(valid_x) == 0:
            raise ValueError("No valid slice interval found.")

        x_l, x_r = valid_x[0], valid_x[-1]

        x = np.random.uniform(x_l, x_r)
        samples[i] = x
    return samples
```

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]


    x_ = a + stats.uniform().rvs()*(b-a)
    if f(x_)>y:
        return x_,1  
    elif x_ < x: 
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_ 
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):

    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)

    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

```python
x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))

m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):

    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')

    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')

plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```
